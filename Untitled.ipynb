## TF IDF  and K -Mean Clustering 

```python
# Text Vectorization
# Term frequency and Inverse Document frequency

vectorizer = TfidfVectorizer(stop_words='english', ngram_range=(1,2))

x = vectorizer.fit_transform(df["title"])
x
```

```python
pd.DataFrame(x.toarray(), columns=vectorizer.get_feature_names())
```

```python
from sklearn.cluster import KMeans


sum_of_squared_distances=[]
# kmeans clustering
K = range(2, 10)
for k in K:
    km = KMeans(n_clusters=k, max_iter=600, n_init=10)
    km.fit(x)
    sum_of_squared_distances.append(km.inertia_)
    
```

```python
plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel("Number of Clusters")
```

```python
# Get Clusters

true_k = 6
model = KMeans(n_clusters = true_k, init='k-means++', max_iter=600, n_init=10)
model.fit(x)

# get prediction/labels
labels = model.labels_
book_c1 = pd.DataFrame(list(zip(df['title'], labels)), columns=['title', 'cluster'])
print(book_c1.sort_values(by=['cluster']))
```

```python
# Create wordclouds for clusters
from wordcloud import WordCloud

for k in range(true_k):
    text = book_c1[book_c1.cluster == k]['title'].str.cat(sep=' ')
    wordcloud = WordCloud(max_font_size=50, max_words=100, background_color='white').generate(text)
    
    
    plt.plot()
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
    
    
```
